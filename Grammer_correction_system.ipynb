```python
!pip install happytransformer # Install the necessary library
from happytransformer import HappyTextToText, TTSettings # import functions from library


happy_tt = HappyTextToText("T5", "vennify/t5-base-grammar-correction")

# Define settings
args = TTSettings(num_beams=5, min_length=1)

# Take user input
user_input = input("Enter a sentence to correct its grammar: ")

# Add the prefix "grammar: " before the input
result = happy_tt.generate_text("grammar: " + user_input, args=args)

# Print the corrected sentence
print("Corrected sentence:", result.text)
```

```python
!pip install happytransformer
!pip install gradio
from happytransformer import HappyTextToText, TTSettings
import gradio as gr

# Initialize the HappyTextToText model
happy_tt = HappyTextToText("T5", "vennify/t5-base-grammar-correction")
args = TTSettings(num_beams=5, min_length=1, max_length=100)

# Define the function for grammar correction
def correct_grammar(text):
    input_text = "grammar: " + text
    result = happy_tt.generate_text(input_text, args=args)
    return result.text

# Create the Gradio interface
# In more recent versions of Gradio, components like Textbox are accessed
# directly from the gradio module.
iface = gr.Interface(
    fn=correct_grammar,
    inputs=gr.Textbox(lines=5, placeholder="Enter text with grammatical errors..."), # Change to gr.Textbox
    outputs="text",
    title="Grammar Corrector",
    description="Enter a sentence or paragraph, and the model will correct its grammar."
)

if __name__ == "__main__":
    iface.launch()
```
